```python
import numpy as np
```

```python
# creating a 1D arry using np.arry()
revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0,55_650.0])

#printing array 
print(revenue)

#print the data type
print(revenue.dtype)

#print the shape 
print(revenue.shape)
```

```python
# Descriptive Statistics

#create the revenue array
revenue =np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0,55_650.0])

# calulate total revenue
total_revenue = np.sum(revenue)

#printing results 
print(total_revenue)
```

```python
# creating revenue array
revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0,55_650.0])

# calulation 
mean_revenue = np.mean(revenue)

#printing results
print(mean_revenue)
```

```python
# creating revenue array
revenue = np.array([52_400.0, 38_750.0, 61_200.0, 47_900.0, 55_650.0])

# calculation
median_revenue = np.median(revenue)

# printing results
print(median_revenue)
```

```python


revenue = np.array([52_400.0, 38_750.0, 61_200.0])

Std_deviation = np.std(revenue)

print(Std_deviation)
```

```python
# finding min and max values
min_value = np.min(revenue)
max_value = np.max(revenue)

#printing results
print(min_value)
print(max_value)
```
