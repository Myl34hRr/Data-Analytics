## Generating Random Numbers

```python
# creating random integers between 1 and 20 with 2 elements
r_i = np.random.randint(1, 20, size=2)

# printing result
print(r_i)
```

```python
# creating random intergers between 1 and 20 with 1 row and 2 columns
r_2 = np.random.randint(1, 20 ,size=(1,2))
# printing result
print (r_2)
```

```python
# creating random floats between 0 and 1 with 2 rows and 1 column
r_3 = np.random.rand(2,1)

# printing result
print(r_3)
```

```python
# creating random intergers between 20 and 100 with 4 rows and 5 columns
r_4 = np.random.randint(20, 100, size=(4,5))

# print reslut
print(r_4)
```

```python
# creating random floats between 0 and 1 with 4 rows and 5 columns
r_5 = np.random.rand(4,5)
print(r_5)
```

## Generating Random Choices from an Array

```python
# create a 1-D array of student names
students = np.array([
    "Ava",
    "Liam",
    "Sophia",
    "Noah",
    "Mia",
    "Elijah",
    "Emma",
    "James",
    "Olivia",
    "Lucas",
    "Charlotte",
    "Benjamin"
])

#print array
print(students)
```

```python
# choosing random student name
call_on = np.random.choice(students)

# print chocie
print(call_on)
```

```python
# choosing random students with 3 row and 4 columns 
breakout_groups = np.random.choice(students,size=(3,4))

#print chocies
print(breakout_groups)
```

```python
# no this wouldn't be an effective way for breakout groups becuase some students come up more than once making the group uneven. 
```

```python
# choosing five random students
call_on_weighted = np.random.choice(students,
                                    size=5,
                                    p=[
    
        0.30,  # Ava has a 30% chance
        0.10,  # Liam has a 10% chance
        0.10,  # Sophia has a 10% chance
        0.08,  # Noah has an 8% chance
        0.07,  # Mia has a 7% chance
        0.07,  # Elijah has a 7% chance
        0.06,  # Emma has a 6% chance
        0.05,  # James has a 5% chance
        0.05,  # Olivia has a 5% chance
        0.05,  # Lucas has a 5% chance
        0.07,  # Charlotte has a 7% chance
        0.00   # Benjamin has a 0% chance (will never be selected)
                                    ]
                                   )
# print choice 
print(call_on_weighted)
```

## Generating a Normal Distribution

```python
# normal (bell curve)distribution
bell_curve = np.random.normal(loc=10, scale=5, size=(4,5))

# printing reslut
print(bell_curve)
```
