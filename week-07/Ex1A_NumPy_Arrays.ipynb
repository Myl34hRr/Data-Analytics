 ## Looping Through Arrays

```python
#creating a 2D array
arrA = np.array([[11,12,13], [14,15,16]])
print(arrA)
```

```python
# Loop through the first dimension of arrA
for x in arrA:
    print(x * 2)
```

```python
# Loop through the second dimension of arrA
for x in arrA:
    for y in x:
        print(y * 2)
```

# the difference between the first and second are the placements in the outputs, the first output loops it virtually while the second output loops it diagonally

```python
# Loop through the second dimension of arrA
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)

# This result is different because x represents the entire row,
# not a single number. Multiplying x by 2 doubles every value
# in the whole row each time the inner loop runs.
```

## Splitting Array

```python
import numpy as np

# Create arrA
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

# Use where() to check for even numbers
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')

# Print result
print(result_1)
```

## Sorting and Filtering Arrays

```python
import numpy as np

# Create arrB
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

# Sort arrB
arrB_sorted = np.sort(arrB)

# Print sorted array
print(arrB_sorted)
```

```python
import numpy as np

# Create arrB
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

# Print arrB
print(arrB)

# Find values less than 7
z = arrB < 7

# Filter arrB using z
arrC = arrB[z]

# Print arrC
print(arrC)
```

```python
import numpy as np

# Create arrB
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

# Create filter for values greater than or equal to 7
y = arrB >= 7

# Filter arrB using y
arrD = arrB[y]

# Print arrD
print(arrD)
```
