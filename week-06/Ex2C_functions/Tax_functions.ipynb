This lab demonstrates how to build Python functions that perform financial calculations using conditional logic. It shows how different withholding codes affect tax rates and how functions can apply rules using if/elif/else statements. It also reinforces testing functions with real input values.

```python
# Create the function
def get_soc_sec_tax(gross_pay):

    # Calculate Social Security tax
    soc_sec_tax = gross_pay * 0.062

    # Return the tax amount
    return soc_sec_tax
```

```python
def get_medicare_tax(gross_pay):

# calucate medicare tax
    medicare_tax = gross_pay * 0.0145

# return the tax amount
    return medicare_tax
```

```python


# calucate federal tax
def get_federal_tax(gross_pay, code):

    # determine tax rate based on withholding code
    if code == 0:
        tax_rate = 0.23

    elif code == 1:
        tax_rate = 0.21

    elif code == 2:
        tax_rate = 0.195

    elif code == 3:
        tax_rate = 0.185

    else:
        # code 4 or higher
        # starts at 18% and subtracts 0.5% per level above 4
        extra_levels = code - 4
        tax_rate = 0.18 - (extra_levels * 0.005)

    # calculate tax
    federal_tax = gross_pay * tax_rate

    return federal_tax
```

```python
print(get_federal_tax(750, 0))
```

```python
print(get_federal_tax(1550, 2))
```

```python
print(get_federal_tax(1100, 5))
```
