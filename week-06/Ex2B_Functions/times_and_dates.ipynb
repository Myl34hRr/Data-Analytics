```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
print(today)
```

```python


formatted_date = today.strftime("Today is %A, %B %d, %Y")

print(formatted_date)
```

# My Birthday

```python
Birthday = datetime.date(2004, 8, 21,)
print(Birthday)
                    
```

```python


formatted_Birthday = Birthday.strftime("My birthday is %m/%d/%Y")

print(formatted_Birthday)
```

# Next 90 Days

```python


today = datetime.datetime.now()

ninety_d = today + datetime.timedelta(days=90)
```

```python
print(ninety_d.strftime("90 days fromt today is %B %d, %Y"))
```

# Time for Dinner

```python
dinner_time = datetime.time()
```

```python
print(dinner_time.strftime("Let's meet for dinner at HH:MM:PM"))
```
